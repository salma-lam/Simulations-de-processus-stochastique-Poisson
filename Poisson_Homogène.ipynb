```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Paramètres
lambda_homogene = 5  # Taux d'événements par unité de temps
temps_total = 10  # Durée de la simulation en unités de temps
```

```python
# Génération des temps inter-arrivées avec une distribution exponentielle
inter_arrivees = np.random.exponential(1 / lambda_homogene, size=1000)
temps_arrivees = np.cumsum(inter_arrivees)
temps_arrivees = temps_arrivees[temps_arrivees <= temps_total]  # Limiter aux événements dans l'intervalle
```

```python
# Visualisation
plt.step(temps_arrivees, np.arange(1, len(temps_arrivees) + 1), where='post')
plt.xlabel('Temps')
plt.ylabel('Nombre d\'événements')
plt.title('Processus de Poisson Homogène')
plt.show()
```
