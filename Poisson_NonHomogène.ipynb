```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Définition du taux variable lambda(t)
def lambda_t(t):
    return 5 + 2 * np.sin(t)  # Exemple de taux variable
```

```python
# Paramètres
lambda_max = 7  # Taux maximal utilisé pour la méthode de rejet
temps_total = 10  # Durée totale de la simulation en unités de temps
```

```python
# Initialisation de la liste des temps d'arrivée pour le processus non homogène
temps_arrivees_non_homogene = []
temps = 0
```

```python
# Simulation par la méthode de rejet
while temps < temps_total:
    # Générer un temps inter-arrivée avec un taux constant lambda_max
    temps += np.random.exponential(1 / lambda_max)
    if temps >= temps_total:
        break
    # Accepter l'événement avec une probabilité proportionnelle à lambda(t) / lambda_max
    if np.random.uniform(0, 1) < lambda_t(temps) / lambda_max:
        temps_arrivees_non_homogene.append(temps)
```

```python

# Visualisation
plt.step(temps_arrivees_non_homogene, np.arange(1, len(temps_arrivees_non_homogene) + 1), where='post')
plt.xlabel('Temps')
plt.ylabel("Nombre d'événements")
plt.title('Processus de Poisson Non Homogène')
plt.show()
```
